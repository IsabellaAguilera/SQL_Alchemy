```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
import matplotlib.pyplot as plt
```

```python
connection_string = "postgres:postgres@localhost:5432/EmployeesSQL"
engine = create_engine(f'postgresql://{connection_string}')
conn=engine.connect()
```

```python
engine.table_names()
```

```python
salaries = pd.read_sql("SELECT * FROM salaries", conn)
salaries.head()
```

```python
titles = pd.read_sql("SELECT * FROM titles", conn)
titles.head()
```

```python
salaries_titles_merge = pd.merge(salaries,titles, on="emp_no", how="inner")
salaries_titles_merge.head()
```

```python
mean_grouped = salaries_titles_merge.groupby("title").mean()
mean_grouped
```

```python
title_salary=mean_grouped.drop(columns="emp_no")
title_salary
```

```python
title_salary=title_salary.reset_index()
title_salary
```

```python
xaxis = title_salary["title"]
yaxis = title_salary["salary"]
ticks = np.arange(len(xaxis))

plt.bar(xaxis, yaxis, align='center', color=["red","blue","yellow","purple","green","orange","lightblue"])

plt.xticks(ticks, xaxis, rotation='vertical')

plt.ylabel("Salaries in Dollars")
plt.xlabel("Job Title")
plt.title("Average Salary by Job Title")

plt.savefig('salary_by_title.png')

plt.show()
```
